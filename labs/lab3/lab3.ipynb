```
import pandas as pd
from sklearn.preprocessing import StandardScaler

```

```
DATA_PATH = "data/data.csv"

df = pd.read_csv(DATA_PATH)

df.info()
df.head()
```

```
df.isna().sum()
```

```
target_col = "activity"

y = df[target_col]
X = df.drop(columns=[target_col])
```

```
X.dtypes.value_counts()
```

```
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# при желании можно вернуть DataFrame с теми же именами признаков
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled_df.head()
```

```
import itertools
import numpy as np

from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture

clusterers_config = [
    {
        "name": "kmeans",
        "estimator": KMeans,
        "param_grid": {
            "n_clusters": list(range(2, 13)),
            "init": ["k-means++", "random"],
            "n_init": [10, 20],
            "random_state": [42],
        },
    },
    {
        "name": "agglomerative",
        "estimator": AgglomerativeClustering,
        "param_grid": {
            "n_clusters": list(range(2, 13)),
            "linkage": ["ward"],
        },
    },
    {
        "name": "dbscan",
        "estimator": DBSCAN,
        "param_grid": {
            "eps": [0.3, 0.5, 0.7, 1.0, 1.3, 1.6, 2.0],
            "min_samples": [3, 5, 10, 20],
        },
    },
    {
        "name": "gmm",
        "estimator": GaussianMixture,
        "param_grid": {
            "n_components": list(range(2, 13)),
            "covariance_type": ["full", "tied"],
            "random_state": [42],
        },
    },
]



def iter_param_grid(param_grid):
    keys = list(param_grid.keys())
    values = [param_grid[k] for k in keys]
    for combo in itertools.product(*values):
        yield dict(zip(keys, combo))


cluster_results = []

for cfg in clusterers_config:
    name = cfg["name"]
    Estimator = cfg["estimator"]
    for params in iter_param_grid(cfg["param_grid"]):
        model = Estimator(**params)

        labels = model.fit_predict(X_scaled)


        unique_labels = np.unique(labels)
        if -1 in unique_labels:
            n_clusters = len(unique_labels) - 1
            n_noise = np.sum(labels == -1)
        else:
            n_clusters = len(unique_labels)
            n_noise = 0

        cluster_results.append({
            "algo": name,
            "params": params,
            "labels": labels,
            "n_clusters": n_clusters,
            "n_noise": n_noise,
        })

        print(f"{name}, params={params}")
        print(f"  n_clusters = {n_clusters}, n_noise = {n_noise}")
        print()
```

```
from sklearn import metrics

internal_results = []

for res in cluster_results:
    labels = res["labels"]
    n_clusters = res["n_clusters"]

    if n_clusters < 2:
        continue

    try:
        sil = metrics.silhouette_score(X_scaled, labels)
        ch = metrics.calinski_harabasz_score(X_scaled, labels)
        db = metrics.davies_bouldin_score(X_scaled, labels)
    except Exception as e:
        print(f"Skip {res['algo']} {res['params']} due to error: {e}")
        continue

    internal_results.append({
        "algo": res["algo"],
        "params": res["params"],
        "n_clusters": n_clusters,
        "n_noise": res["n_noise"],
        "silhouette": sil,
        "calinski_harabasz": ch,
        "davies_bouldin": db,
    })

len(internal_results)
```

```
internal_df = pd.DataFrame(internal_results)

internal_df.describe()
```

```
internal_df.sort_values("silhouette", ascending=False)
```

```
best_row = internal_df.sort_values("silhouette", ascending=False).iloc[0]

best_algo_name = best_row["algo"]
best_params = best_row["params"]
best_n_clusters = best_row["n_clusters"]

print("Лучшая конфигурация по силуэту:")
print("  algo      =", best_algo_name)
print("  params    =", best_params)
print("  n_clusters=", best_n_clusters)
print("  silhouette=", best_row["silhouette"])
print("  CH        =", best_row["calinski_harabasz"])
print("  DB        =", best_row["davies_bouldin"])
```

```
best_cfg = next(cfg for cfg in clusterers_config if cfg["name"] == best_algo_name)
BestEstimator = best_cfg["estimator"]

best_model = BestEstimator(**best_params)
best_labels = best_model.fit_predict(X_scaled)

print("unique labels:", np.unique(best_labels))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "cluster": best_labels,
})

plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="cluster",
    palette="tab10",
    s=15,
    alpha=0.8,
    edgecolor="none",
)
plt.title(f"PCA-проекция, кластеры ({best_algo_name}, {best_n_clusters} кластеров)")
plt.legend(title="Кластер", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```
pca_true_df = pca_df.copy()
pca_true_df["activity"] = y.values

plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=pca_true_df,
    x="PC1",
    y="PC2",
    hue="activity",
    s=15,
    alpha=0.8,
    edgecolor="none",
)
plt.title("PCA-проекция, истинные активности")
plt.legend(title="activity", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```
from sklearn import metrics

ari_best = metrics.adjusted_rand_score(y, best_labels)
ami_best = metrics.adjusted_mutual_info_score(y, best_labels)
hom_best = metrics.homogeneity_score(y, best_labels)
comp_best = metrics.completeness_score(y, best_labels)
v_best = metrics.v_measure_score(y, best_labels)

print("Внешние метрики для лучшей конфигурации:")
print(f"  ARI         = {ari_best:.4f}")
print(f"  AMI         = {ami_best:.4f}")
print(f"  Homogeneity = {hom_best:.44f}")
print(f"  Completeness= {comp_best:.4f}")
print(f"  V-measure   = {v_best:.4f}")
```

```
n_classes = y.nunique()
print(n_classes, y.value_counts())
```

```
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn import metrics

fixed_clusterers = [
    ("kmeans_6", KMeans(n_clusters=n_classes, init="k-means++", n_init=20, random_state=42)),
    ("agglomerative_6", AgglomerativeClustering(n_clusters=n_classes, linkage="ward")),
    ("gmm_6", GaussianMixture(n_components=n_classes, covariance_type="full", random_state=42)),
]

fixed_results = []

for name, model in fixed_clusterers:
    labels = model.fit_predict(X_scaled)

    ari = metrics.adjusted_rand_score(y, labels)
    ami = metrics.adjusted_mutual_info_score(y, labels)
    hom = metrics.homogeneity_score(y, labels)
    comp = metrics.completeness_score(y, labels)
    v   = metrics.v_measure_score(y, labels)

    fixed_results.append({
        "name": name,
        "labels": labels,
        "ARI": ari,
        "AMI": ami,
        "Homogeneity": hom,
        "Completeness": comp,
        "V_measure": v,
    })

fixed_df = pd.DataFrame(fixed_results)[
    ["name", "ARI", "AMI", "Homogeneity", "Completeness", "V_measure"]
]
fixed_df
```

```
best_fixed = max(fixed_results, key=lambda d: d["ARI"])
best_fixed_name = best_fixed["name"]
best_fixed_labels = best_fixed["labels"]

print("Лучшая модель при k = числу классов:")
print(best_fixed_name)
print(f"ARI = {best_fixed['ARI']:.4f}, AMI = {best_fixed['AMI']:.4f}")
print(f"Homogeneity = {best_fixed['Homogeneity']:.4f}, Completeness = {best_fixed['Completeness']:.4f}")
print(f"V-measure   = {best_fixed['V_measure']:.4f}")
```

```
crosstab_k_equal = pd.crosstab(y, best_fixed_labels)
crosstab_k_equal
```

```
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics


k_values = list(range(2, 21))

inertias = []
silhouettes = []

for k in k_values:
    km = KMeans(
        n_clusters=k,
        init="k-means++",
        n_init=20,
        random_state=42
    )
    labels = km.fit_predict(X_scaled)

    inertias.append(km.inertia_)

    sil = metrics.silhouette_score(X_scaled, labels)
    silhouettes.append(sil)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(k_values, inertias, marker="o")
ax[0].set_xlabel("Число кластеров k")
ax[0].set_ylabel("Inertia (SSE)")
ax[0].set_title("KMeans: метод локтя")

ax[1].plot(k_values, silhouettes, marker="s")
ax[1].set_xlabel("Число кластеров k")
ax[1].set_ylabel("Silhouette score")
ax[1].set_title("KMeans: силуэт vs k")

plt.tight_layout()
plt.show()
```
